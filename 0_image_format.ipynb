### Step 1: Functions

```python
def check_out_dir(out_dir, item):
  ## create if not exist
  if not os.path.exists(os.path.join(OUT_DIR, item)):
    os.makedirs(os.path.join(OUT_DIR, item))
  ## clean folder if exist
  else:
    files = os.listdir(os.path.join(OUT_DIR, item))
    for f in files:
      os.remove(os.path.join(OUT_DIR, item, f))

def convert_format(in_dir, out_dir, item):
  images = os.listdir(os.path.join(in_dir, item))
  for _, image in enumerate(images):
    ## format heic 
    if (image.split('.')[1].upper() == 'HEIC'):
      heif_file = pillow_heif.read_heif(os.path.join(in_dir, item, image))
      img = Image.frombytes(
          heif_file.mode,
          heif_file.size,
          heif_file.data,
          "raw")
      img.save(os.path.join(out_dir, item, image.split('.')[0] + '.JPG'))
    ## other non-jpg format
    elif (image.split('.')[1].upper() != 'JPG'):
      img = Image.open(os.path.join(in_dir, item, image))
      img_rgb = img.convert('RGB')
      img_rgb.save(os.path.join(out_dir, item, image.split('.')[0] + '.JPG'))
    ## jpg format
    else:
      _ = shutil.copy(os.path.join(IN_DIR, item, image), os.path.join(OUT_DIR, item, image))

def check_folder_size(in_dir, out_dir, item):
  num_in = len(os.listdir(os.path.join(in_dir, item)))
  num_out = len(os.listdir(os.path.join(out_dir, item)))
  print(f'   transfer {num_in} source files to {num_out} at destination   @   {dtdt.now()}')
```
